```python
import pandas as pd
```

```python
df = pd.read_csv('respuestas.csv')
df
```

```python
coordenadas=[]
```

```python
for index, row in df.iterrows():
  lat = row['latitud']
  lon = row['longitud']
  coordenadas.append(str(lat)+', '+str(lon))
```

```python
coordenadas
```

```python
df['coordenadas'] = coordenadas
df
```

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
geolocator = Nominatim(user_agent = 'mooc_igic')
reverse = RateLimiter(geolocator.reverse, min_delay_seconds=1)
df['ubicacion'] = df['coordenadas'].apply(reverse)
df['Direccion'] = df['ubicacion'].apply(lambda x: (x.address))
df
```

```python
df.to_csv('direcciones.csv')
```
